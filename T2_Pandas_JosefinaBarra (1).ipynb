```python
species_df = pd.read_csv('species.csv')
species_df.head()
```

```python
union_df = pd.merge(surveys_df, species_df, on='species_id')
union_df.head()
```

```python
union_df['taxa'] = union_df['genus'] + ' ' + union_df['species']

taxa_by_plot = union_df.groupby(['plot_id', 'taxa'])['record_id'].count()

taxa_by_plot
```

```python
taxa_sex_plot = union_df.groupby(['plot_id', 'taxa', 'sex'])['record_id'].count()
taxa_sex_plot
```

```python
plots_df = pd.read_csv('plots.csv')
plots_df.head()
```

```python
tipos_parcelas = plots_df['plot_type'].value_counts()
tipos_parcelas
```

```python
counts_df = surveys_df.groupby(['plot_id', 'species_id'])['record_id'].count().reset_index()
counts_df = counts_df.rename(columns={'record_id': 'count'})

unique_species = counts_df.groupby('plot_id')['species_id'].nunique().reset_index()
unique_species = unique_species.rename(columns={'species_id': 'num_species'})

total_counts = counts_df.groupby('plot_id')['count'].sum().reset_index()

biodiv_df = pd.merge(total_counts, unique_species, on='plot_id')

biodiv_df['biodiversity_index'] = biodiv_df['num_species'] / biodiv_df['count']

biodiv_df = biodiv_df[biodiv_df['plot_id'].isin(['Control', 'Rodent Exclosure'])]

print(biodiv_df)
```
