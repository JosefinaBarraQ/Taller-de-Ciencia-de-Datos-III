Importar las siguientes librer√≠as: `matplotlib.pyplot, seaborn, pandas, numpy`

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
```

## <b>Recrea la siguiente figura</b>

```python
x = np.linspace(-4, 4, 52)

# Calcula valores para curvas
y1 = 3 * np.exp(-0.25 * x**2)
y2 = 2.8 * np.exp(-0.15 * (x - 0.1)**2)

# Configura estilo de seaborn
sns.set(style="darkgrid")

# Crea figura y ejes
fig, ax = plt.subplots()

# Curvas
sns.lineplot(x, y1, color='red', linestyle='-.', linewidth=2, label='3e^(-0.25x^2)', ax=ax)
sns.lineplot(x, y2, color='cyan', marker='o', markerfacecolor='yellow', markeredgecolor='black', markersize=6, linestyle='-', linewidth=2, label='2.8e^(-0.15(x-0.1)^2)', ax=ax)

# Etiquetas
ax.set_xlabel("That took a while", fontweight='bold')
ax.set_title('Many Plots', fontweight='bold')
ax.legend(loc='upper right')

ax2 = plt.axes([0.41, 0.18, 0.2, 0.3])
ax2.text(-1.5, 2, "Cool", color='red')
ax2.text(-0.5, 0, "Plot", color='blue')
ax2.text(1, -2, "Bro", color='green')

ax2.set_title('Small Plot')
ax2.set_xlim(-2, 2)
ax2.set_ylim(-3, 3)
ax2.set_xticks(np.arange(-2, 2.5, 1))
ax2.set_yticks(np.arange(-3, 3.5, 1))

plt.show()

```

## <b> Recrea el siguiente  `violinplot` comparando la distribuci√≥n de  `Entire home/apt` y `Private room` para los 5 `neighbourhood groups`</b>

```python
filtered_abnb = abnb.loc[abnb['room_type'].isin(['Entire home/apt', 'Private room'])]

sns.set(style="ticks")

plt.figure(figsize=(10, 6))

sns.catplot(data=filtered_abnb, x='neighbourhood_group', y='price', hue='room_type', kind='violin', split=True)


plt.xlabel('Neighbourhood Group')
plt.ylabel('Price')

plt.show()
```


## <b>Time Series: Recrea el siguiente gr√°fico</b>

(Pista: Convierte la columna `last_review` a formato `DateTime`)

```python
#No logre terminar este ejercicio, quedo muy raro
abnb['last_review'] = pd.to_datetime(abnb['last_review'])


start_date = pd.to_datetime('2016-10-01')
end_date = pd.to_datetime('2019-03-01')
abnb_timepo = abnb.loc[(abnb['last_review'] >= start_date) & (abnb['last_review'] < end_date)]

sns.set(style="ticks")


plt.figure(figsize=(10, 6))

plt.plot(abnb_timepo['last_review'], abnb_timepo['number_of_reviews'], label='Number of Reviews')
plt.plot(abnb_timepo['last_review'], abnb_timepo['price'], label='Price')

plt.xlabel('last_review')
plt.ylabel('')
plt.title('Fluctuations in Number of Reviews and Price over time')

plt.legend()
plt.show()
```
